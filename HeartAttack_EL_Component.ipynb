Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Reading the Dataset

```python
ds=pd.read_csv("PLC EL DATASET.csv")
ds.shape
```

```python
ds.head()
```

```python
ds['sex'].replace({1:'Male',0:'Female'}, inplace=True)
ds['target'].replace({0:'No Disease',1:'Disease'},inplace=True)
ds['cp'].replace({ 0:'Typical angina',1:'Atypical angina',
                  2: 'Non-anginal pain',3: 'Asymptomatic'},inplace=True)
ds['slope'].replace({0:'Upsloping',1:'Flatsloping',
                     2:'Downsloping'},inplace=True)
ds['fbs'].replace({0:'Abnormal',1:'Normal'},inplace=True)
ds['exang'].replace({0:'No',1:'Yes'},inplace=True)
ds['restecg'].replace({0:'Abnormal',1:'Normal'},inplace=True)

ds.head()
```

Training and testing data

```python
ds=pd.read_csv("PLC EL DATASET.csv")
ds.shape
```

```python
X=ds.drop('target',axis=1)# axis=1: Drop the specified column(s) (along columns), axis=0: Drop the specified row(s) (along rows).
y=ds.target.values
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
X_train
```

```python
y_train
```

Implementation of Logistic Regression

```python
regressor=LogisticRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)# train() test() pred()       Math sets Exercise u hve 20(back answer),5 miscc (back answer)
                                                                                      #15 questions trains, 5 test, 5 Pred
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
